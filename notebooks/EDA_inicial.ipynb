```python
#Identificación de patrones entre variables para detección de operadores ineficientes 
resumen_operadores = df2.groupby("operator_id").agg(
    total_llamadas = ("calls_count", "sum"),
    llamadas_perdidas = ("is_missed_call", "sum"),
    llamadas_salientes = ("direction", lambda x: (x=="out").sum()),
    duracion_promedio = ("call_duration", "mean"),
    duracion_total = ("total_call_duration", "sum")
).reset_index()

resumen_operadores
```

```python
#% de llamadas perdidas 
resumen_operadores["porcentaje_perdidas"] = (
    resumen_operadores["llamadas_perdidas"] / resumen_operadores["total_llamadas"] * 100
)

resumen_operadores.head(20)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    resumen_operadores["total_llamadas"],
    resumen_operadores["porcentaje_perdidas"],
    alpha=0.5
)
plt.xlabel("Total de llamadas")
plt.ylabel("% de llamadas perdidas")
plt.title("Relación entre volumen de llamadas y % de pérdidas")
plt.show()
```

Con este gráfico podemos interpretar que a medida que aumenta el volumen total de llamadas el porcentaje de perdidas tiende a disminuir de forma significativa, podemos deducir que existe una gran diferencia ya que entre mayor sea el número de llamadas existe una menor tasa de llamadas perdidas.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=resumen_operadores,
    x="duracion_promedio",
    y="porcentaje_perdidas",
    alpha=0.6
)
plt.xlabel("Duración promedio de llamadas (segundos)")
plt.ylabel("% de llamadas perdidas")
plt.title("Duración promedio vs % de pérdidas")
plt.show()
```

En este gráfico podemos observar que cuando las llamadas cuentan con una duración promedio muy corta, hay una variabilidad muy grande en la cantidad de perdidas (siendo que cuando la duración promedio es 0 significa que no están completando llamadas exitosas), como en algunos casos donde se llega al 100% y con forme aumenta la diuración de las llamadas se estabiliz

```python
resumen_operadores["porcentaje_perdidas"] = (
    resumen_operadores["llamadas_perdidas"] / resumen_operadores["total_llamadas"] * 100
)

top_perdedores = resumen_operadores.sort_values("porcentaje_perdidas", ascending=False).head(20)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.barh(top_perdedores["operator_id"].astype(str), top_perdedores["porcentaje_perdidas"], color="tomato")
plt.xlabel("% de llamadas perdidas")
plt.ylabel("Operator ID")
plt.title("Top 20 operadores con mayor % de llamadas perdidas")
plt.gca().invert_yaxis()
plt.show()
```

En este gráfico se ha desidido visualizar un rango de los elementos menos eficientes, esto muestra un factor preocupante ya que se muestra que tienen un 100% de llamadas perdidas, lo cual podría indicar una ineficiencia total. Es importamnte tomar en cuenta rangos de tiempo o algún otro factor que pueda ser un factor que provoque este resultado.
