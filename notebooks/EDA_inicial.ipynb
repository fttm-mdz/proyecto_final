```python
#Identificación de patrones entre variables para detección de operadores ineficientes 
resumen_operadores = df2.groupby("operator_id").agg(
    total_llamadas = ("calls_count", "sum"),
    llamadas_perdidas = ("is_missed_call", "sum"),
    llamadas_salientes = ("direction", lambda x: (x=="out").sum()),
    duracion_promedio = ("call_duration", "mean"),
    duracion_total = ("total_call_duration", "sum")
).reset_index()

resumen_operadores
```

Este resumen proporciona una visión general del desempeño de cada operador, incluyendo el total de llamadas, llamadas perdidas, llamadas salientes, duración promedio de las llamadas y duración total. Esto nos permite identificar a los operadores con mejor y peor desempeño en cada una de estas métricas. Es importante destacar a los operadores con un alto número de llamadas perdidas, ya que esto podría indicar problemas de eficiencia o falta de recursos.


```python
#% de llamadas perdidas 
resumen_operadores["porcentaje_perdidas"] = (
    resumen_operadores["llamadas_perdidas"] / resumen_operadores["total_llamadas"] * 100
)

resumen_operadores.head(20)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    resumen_operadores["total_llamadas"],
    resumen_operadores["porcentaje_perdidas"],
    alpha=0.5
)
plt.xlabel("Total de llamadas")
plt.ylabel("% de llamadas perdidas")
plt.title("Relación entre volumen de llamadas y % de pérdidas")
plt.show()
```

Este gráfico muestra la relación entre el volumen total de llamadas y el porcentaje de llamadas perdidas para cada operador. En general, se observa una tendencia a que los operadores con un menor volumen de llamadas tiendan a tener un mayor porcentaje de llamadas perdidas, mientras que los operadores con un mayor volumen de llamadas tienden a tener un menor porcentaje de llamadas perdidas. Es importante identificar a los operadores que se desvían de esta tendencia general, ya que podrían ser outliers que requieren atención especial. La tendencia observada podría explicarse por varios factores. Los operadores con un menor volumen de llamadas podrían estar dedicando más tiempo a cada llamada, lo que reduce su disponibilidad para atender nuevas llamadas.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=resumen_operadores,
    x="duracion_promedio",
    y="porcentaje_perdidas",
    alpha=0.6
)
plt.xlabel("Duración promedio de llamadas (segundos)")
plt.ylabel("% de llamadas perdidas")
plt.title("Duración promedio vs % de pérdidas")
plt.show()
```

Se observa una alta variabilidad en el porcentaje de llamadas perdidas cuando la duración promedio es muy corta, lo que sugiere que estos operadores podrían estar teniendo dificultades para establecer una conexión exitosa con los clientes. La alta variabilidad en el porcentaje de llamadas perdidas cuando la duración promedio es muy corta podría deberse a varios factores, como problemas técnicos, errores de marcación, falta de disponibilidad del cliente o una gestión ineficiente de las llamadas. Es importante investigar las razones detrás de esta disparidad en el volumen de llamadas entre los operadores.

```python
#mejores operadores
mejores = resumen_operadores[(resumen_operadores["porcentaje_perdidas"] <= 5) & (resumen_operadores["total_llamadas"] >= 10)]

#peores operadores
peores = resumen_operadores[resumen_operadores["porcentaje_perdidas"] == 100]

#extraer series para comparación
duracion_mejores = mejores["duracion_promedio"]
duracion_peores = peores["duracion_promedio"]

llamadas_mejores = mejores["total_llamadas"]
llamadas_peores = peores["total_llamadas"]

print("Duración promedio de llamadas (mejores operadores):\n", duracion_mejores)
print("\nDuración promedio de llamadas (peores operadores):\n", duracion_peores)
print("\nTotal de llamadas (mejores operadores):\n", llamadas_mejores)
print("\nTotal de llamadas (peores operadores):\n", llamadas_peores)
```

```python
from scipy.stats import ttest_ind

#comparación de duración promedio
t_stat_dur, p_val_dur = ttest_ind(duracion_mejores, duracion_peores, equal_var=False)
print("Duración promedio de llamadas:")
print(f"t-statistic: {t_stat_dur:.2f}, p-value: {p_val_dur:.4f}")

#comparación de total de llamadas
t_stat_ll, p_val_ll = ttest_ind(llamadas_mejores, llamadas_peores, equal_var=False)
print("\nTotal de llamadas:")
print(f"t-statistic: {t_stat_ll:.2f}, p-value: {p_val_ll:.4f}")
```

Nuestros hallazgos proporcionan evidencia estadística que sugiere que los operadores con un menor porcentaje de llamadas perdidas tienden a completar un mayor número de llamadas y a mantener conversaciones más extensas. Esto apoya la hipótesis de que la eficiencia de un operador está positivamente correlacionada con la duración de las llamadas y el volumen de atención brindada.
