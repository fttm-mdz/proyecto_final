```python
#Identificación de patrones entre variables para detección de operadores ineficientes 
resumen_operadores = df2.groupby("operator_id").agg(
    total_llamadas = ("calls_count", "sum"),
    llamadas_perdidas = ("is_missed_call", "sum"),
    llamadas_salientes = ("direction", lambda x: (x=="out").sum()),
    duracion_promedio = ("call_duration", "mean"),
    duracion_total = ("total_call_duration", "sum")
).reset_index()

resumen_operadores
```

```python
#% de llamadas perdidas 
resumen_operadores["porcentaje_perdidas"] = (
    resumen_operadores["llamadas_perdidas"] / resumen_operadores["total_llamadas"] * 100
)

resumen_operadores.head(20)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    resumen_operadores["total_llamadas"],
    resumen_operadores["porcentaje_perdidas"],
    alpha=0.5
)
plt.xlabel("Total de llamadas")
plt.ylabel("% de llamadas perdidas")
plt.title("Relación entre volumen de llamadas y % de pérdidas")
plt.show()
```

Con este gráfico podemos interpretar que a medida que aumenta el volumen total de llamadas el porcentaje de perdidas tiende a disminuir de forma significativa, podemos deducir que existe una gran diferencia ya que entre mayor sea el número de llamadas existe una menor tasa de llamadas perdidas.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=resumen_operadores,
    x="duracion_promedio",
    y="porcentaje_perdidas",
    alpha=0.6
)
plt.xlabel("Duración promedio de llamadas (segundos)")
plt.ylabel("% de llamadas perdidas")
plt.title("Duración promedio vs % de pérdidas")
plt.show()
```

En este gráfico podemos observar que cuando las llamadas cuentan con una duración promedio muy corta, hay una variabilidad muy grande en la cantidad de perdidas (siendo que cuando la duración promedio es 0 significa que no están completando llamadas exitosas), como en algunos casos donde se llega al 100% y con forme aumenta la diuración de las llamadas se estabiliz

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=resumen_operadores, y="total_llamadas")
plt.title("Distribución del total de llamadas por operador")
plt.ylabel("Total de llamadas")
plt.show()

```

La mayoría de los operadores tienen un número relativamente bajo de llamadas, como lo indica la caja compacta cerca de la parte inferior del gráfico. Sin embargo, existen varios valores atípicos que representan a operadores con un volumen de llamadas significativamente mayor. Estos valores atípicos sugieren que algunos operadores son responsables de una gran proporción del total de llamadas realizadas. Es importante investigar las razones detrás de esta disparidad en el volumen de llamadas entre los operadores.

```python
#mejores operadores
mejores = resumen_operadores[(resumen_operadores["porcentaje_perdidas"] <= 5) & (resumen_operadores["total_llamadas"] >= 10)]

#peores operadores
peores = resumen_operadores[resumen_operadores["porcentaje_perdidas"] == 100]

#extraer series para comparación
duracion_mejores = mejores["duracion_promedio"]
duracion_peores = peores["duracion_promedio"]

llamadas_mejores = mejores["total_llamadas"]
llamadas_peores = peores["total_llamadas"]

print("Duración promedio de llamadas (mejores operadores):\n", duracion_mejores)
print("\nDuración promedio de llamadas (peores operadores):\n", duracion_peores)
print("\nTotal de llamadas (mejores operadores):\n", llamadas_mejores)
print("\nTotal de llamadas (peores operadores):\n", llamadas_peores)
```

```python
from scipy.stats import ttest_ind

#comparación de duración promedio
t_stat_dur, p_val_dur = ttest_ind(duracion_mejores, duracion_peores, equal_var=False)
print("Duración promedio de llamadas:")
print(f"t-statistic: {t_stat_dur:.2f}, p-value: {p_val_dur:.4f}")

#comparación de total de llamadas
t_stat_ll, p_val_ll = ttest_ind(llamadas_mejores, llamadas_peores, equal_var=False)
print("\nTotal de llamadas:")
print(f"t-statistic: {t_stat_ll:.2f}, p-value: {p_val_ll:.4f}")
```

Nuestros hallazgos proporcionan evidencia estadística que sugiere que los operadores con un menor porcentaje de llamadas perdidas tienden a completar un mayor número de llamadas y a mantener conversaciones más extensas. Esto apoya la hipótesis de que la eficiencia de un operador está positivamente correlacionada con la duración de las llamadas y el volumen de atención brindada.

```python
#comparación visual de los mejores y peores operadores
plt.figure(figsize=(8,6))
sns.boxplot(
    x="grupo", 
    y="duracion_promedio",
    data=pd.concat([mejores.assign(grupo="Mejor"), operadores_100.assign(grupo="Peor")])
)
plt.title("Duración promedio de llamadas: Mejores vs Peores operadores")
plt.ylabel("Duración promedio (segundos)")
plt.xlabel("Grupo de operadores")
plt.show()
```

El grupo 'Mejor' exhibe una duración media de llamada de aproximadamente 600-700 segundos, con un rango intercuartílico sustancial que indica una amplia distribución de la duración de las llamadas entre estos operadores. La presencia de numerosos valores atípicos, que se extienden más allá de los 14.000 segundos, destaca casos de llamadas excepcionalmente largas, que posiblemente reflejan la resolución de problemas complejos o escenarios de soporte especializado. En contraste, el grupo 'Peor' demuestra una distribución muy comprimida, con el diagrama de caja colapsando cerca de cero. Esto sugiere duraciones de llamada mínimas y consistentes, con una mediana que se acerca a cero segundos, lo que indica un patrón frecuente de interacciones infructuosas o extremadamente breves. La ausencia de valores atípicos notables en este grupo refuerza la uniformidad del bajo rendimiento. Estos hallazgos visuales corroboran la diferencia estadísticamente significativa identificada a través de la prueba t, lo que respalda aún más la conclusión de que los operadores con mejor desempeño participan en interacciones con los clientes sustancialmente más largas y sostenidas en comparación con sus homólogos con peor desempeño, cuyas llamadas se caracterizan por la brevedad y una probable prevalencia de conexiones perdidas.

```python
top_20 = resumen_operadores.nlargest(20, "total_llamadas").index
bottom_20 = resumen_operadores.nsmallest(20, "total_llamadas").index

resumen_operadores["grupo"] = "Medio"
resumen_operadores.loc[top_20, "grupo"] = "Mejor"
resumen_operadores.loc[bottom_20, "grupo"] = "Peor"
```

```python
#resumen comparativo por grupo
resumen_grupos = resumen_operadores.groupby("grupo").agg({
    "duracion_promedio": ["mean", "median"],
    "total_llamadas": ["mean", "median"],
    "llamadas_perdidas": ["mean", "median"]
}).round(2)

resumen_grupos
```

```python
filtros = resumen_operadores[resumen_operadores["grupo"].isin(["Mejor", "Peor"])]

plt.figure(figsize=(12,5))

# Subplot 1 - Duración promedio
plt.subplot(1,3,1)
sns.barplot(data=filtros, x="grupo", y="duracion_promedio", estimator="mean", errorbar=None)
plt.title("Duración Promedio de Llamadas")

# Subplot 2 - Total de llamadas
plt.subplot(1,3,2)
sns.barplot(data=filtros, x="grupo", y="total_llamadas", estimator="mean", errorbar=None)
plt.title("Total de Llamadas")

# Subplot 3 - Llamadas perdidas
plt.subplot(1,3,3)
sns.barplot(data=filtros, x="grupo", y="llamadas_perdidas", estimator="mean", errorbar=None)
plt.title("Llamadas Perdidas")

plt.tight_layout()
plt.show()
```

Los gráficos de barras comparan directamente la Duración Promedio de Llamadas, el Total de Llamadas y las Llamadas Perdidas entre los grupos de operadores 'Mejor' y "Peor". El grupo "Mejor' exhibe una duración promedio significativamente mayor (media = 4182.61 segundos, mediana = 2495.96 segundos) y un número total de llamadas sustancialmente mayor (media = 19874.25, mediana = 6814.5) en comparación con el grupo 'Peor' (Duración Promedio: media = 132.85 segundos, mediana = 25.00 segundos; Total de Llamadas: media = 1.00, mediana = 1.0). Si bien el grupo 'Mejor' experimenta algunas llamadas perdidas (media = 457.1, mediana = 48.0), los valores cercanos a cero del grupo 'Peor' reflejan la falta de conexiones exitosas. En resumen, los operadores 'Mejor' se caracterizan por interacciones más largas, frecuentes y exitosas, mientras que los operadores 'Peor' tienen dificultades para establecer conexiones significativas.
